### Fetch(small test)

```python
from src.youtube_fetcher import YouTubeFetcher

f = YouTubeFetcher(api_key=API_KEY)
meta = f.search_and_fetch("guitar chord lesson beginner intermediate advanced", max_results=2000)
len(meta)
```

### Save & Preview

```python
import pandas as pd

df = pd.DataFrame(meta)
df.head()
# Remove rows with the values of unknown, they'll be saved for the test dataset
test_df = df.query('level == "unknown"').copy()
test_df.to_csv("../data/raw/guitar_sample_test_3.csv")
df = df.drop(df.query("level == 'unknown'").index)
df.to_parquet("../data/raw/guitar_sample_parquet_2", index=False)
```
